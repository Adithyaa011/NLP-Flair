```python
from flair.models import SequenceTagger, TextClassifier
from flair.data import Sentence
```

```python
# Load part-of-speech tagging model
pos_tagger = SequenceTagger.load('pos')

# Load named entity recognition (NER) model
ner_tagger = SequenceTagger.load('ner')

# Load sentiment analysis model
sentiment_classifier = TextClassifier.load('en-sentiment')


# Sample text for NLP tasks
text = "Hello, so I am Aditya from College of Engineering in India and I am recording this for my NLP activity too. So I have to talk for 3 minutes and I have no clue. So I thought I will just tell about my favourite frictional character.Thank you."

# Create a Sentence object
sentence = Sentence(text)


# Perform part-of-speech tagging
pos_tagger.predict(sentence)
pos_tags = sentence.to_tagged_string()


# Perform named entity recognition (NER)
ner_tagger.predict(sentence)
ner_entities = sentence.to_dict(tag_type='ner')


# Perform sentiment analysis
sentiment_classifier.predict(sentence)
sentiment_label = sentence.labels[-1:]





# Display results
print("Original Text:", text)
print()
print("Part-of-Speech Tagging:", pos_tags)
print()
print("Named Entity Recognition (NER):", ner_entities)
print()
print("Sentiment Analysis:", sentiment_label)
```
